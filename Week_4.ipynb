## Importing libraries (Numpy, pandas, matplotlib, plotly and seaborn)

```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns


```

# Calculating the maximum value of the attributes

```python
# Read CSV file
df = pd.read_csv('menu.csv')

#Check statistical facts by checking all columns
statistical_facts = df.describe()

# Calculate the maximum value of the specified attributes
attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein',
              'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)',
              'Calcium (% Daily Value)', 'Iron (% Daily Value)']

max_values = df[attributes].max()

print("Statistical Facts:")
print(statistical_facts)
print("\nMaximum Values:")
print(max_values)
```

## Checking if there is any correlation between Calories and other independent variables by plotting a correlation matrix next.

```python
# Select attributes for correlation matrix
attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein',
              'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)',
              'Calcium (% Daily Value)', 'Iron (% Daily Value)']

# Calculate correlation matrix
correlation_matrix = df[attributes].corr()

# Plot correlation matrix using seaborn with warm-toned colors and without displaying values
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=False, cmap='magma',  linewidths=1.5, linecolor='white', square=True)
plt.title('Correlation Matrix of Nutritional Attributes')
plt.show()

```

## Draw boxplot for Calories vs Category to spot outliers and max calories categor

```python
# Draw horizonta# Create a dictionary to map category names to colors
category_colors = {category: sns.color_palette()[i] for i, category in enumerate(df['Category'].unique())}

# Draw horizontal boxplot for Calories vs Category with legend, major grid lines, and custom legend
plt.figure(figsize=(12, 8))
sns.boxplot(x='Calories', y='Category', data=df, orient='h', palette=category_colors, dodge=False)
plt.title('Boxplot of Calories by Category')
plt.xlabel('Calories')
plt.ylabel('Category')

# Customize legend
handles = [plt.Rectangle((0,0),1,1, color=category_colors[category]) for category in df['Category'].unique()]
plt.legend(handles, df['Category'].unique(), title='Category', loc='lower right')

# Customize major grid lines
plt.grid(which='major', linestyle='-', linewidth='0.5', color='gray')
plt.show()

```

## Seeing the exact item that contains a high quantity for
['Calories', 'Total Fat', 'Carbohydrates','Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium
 (% Daily Value)', 'Iron (% Daily Value)'.

```python
# Dictionary to store the maximum values and corresponding items
max_items = {}

# Find the item with the maximum value for each attribute
for attribute in attributes:
    max_value_index = df[attribute].idxmax()
    max_item_name = df.loc[max_value_index, 'Item']
    max_value = df.loc[max_value_index, attribute]
    max_items[attribute] = {'Item': max_item_name, 'Value': max_value}

# Display the summary
print("Summary:")
for attribute, info in max_items.items():
    print(f"{attribute}: {info['Item']} - {info['Value']}")
```

## Drawing Stripplot

```python
# Define attributes and categories
attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein',
              'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)',
              'Iron (% Daily Value)']

categories = ['Breakfast', 'Beef & Pork', 'Chicken & Fish', 'Salads', 'Snacks & Sides',
              'Desserts', 'Beverages', 'Coffee & Tea', 'Shakes']

# Define custom colors for each category
category_colors = {
    'Breakfast': 'blue',
    'Beef & Pork': 'orange',
    'Chicken & Fish': 'green',
    'Salads': 'red',
    'Snacks & Sides': 'purple',
    'Desserts': 'brown',
    'Beverages': 'pink',
    'Coffee & Tea': 'grey',
    'Smoothies & Shakes': 'limegreen'
}

# Define y-axis ticks
y_ticks = [0, 250, 500, 750, 1000, 1250, 1500, 1750]

# Draw stripplot for each category against the specified attributes
plt.figure(figsize=(30, 10))
for i, attribute in enumerate(attributes, start=1):
    plt.subplot(2, 5, i)
    sns.stripplot(x='Category', y=attribute, data=df, palette=category_colors, jitter=True)
    plt.title(attribute)
    plt.xlabel('')
    plt.ylabel(attribute)
    plt.yticks(y_ticks)
    plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()



```

## Drawing a horizontal bar graph for items in each category against the calories.

```python
# Define categories
categories = ['Breakfast', 'Beef & Pork', 'Chicken & Fish', 'Salads', 'Snacks & Sides',
              'Desserts', 'Beverages', 'Coffee & Tea', 'Smoothies & Shakes']

# Iterate over each category
for category in categories:
    # Filter DataFrame for items in the current category
    category_df = df[df['Category'] == category]

    # Sort items by calories
    category_df = category_df.sort_values(by='Calories', ascending=True)

    # Plot horizontal bar graph
    plt.figure(figsize=(10, len(category_df)*0.5))
    plt.barh(category_df['Item'], category_df['Calories'], color='skyblue')
    plt.title(f'Calories of Items in {category}')
    plt.xlabel('Calories')
    plt.ylabel('Item')
    plt.gca().invert_yaxis()  # Invert y-axis to display items from top to bottom
    # Adjust layout after all plots have been created
    plt.tight_layout()
    plt.show()

#Observation:Observation: By visually inspecting the horizontal bar graphs, we can observe the distribution of calories among items in each category.
#There are variations in calorie content within the same category, indicating differences in portion sizes or ingredients among items.
```
